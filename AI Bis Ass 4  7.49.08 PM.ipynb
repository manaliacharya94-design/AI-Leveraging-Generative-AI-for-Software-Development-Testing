```python
class FactorialCalculator:
    """Calculates the factorial of a non-negative integer."""

    def __repr__(self):
        return "FactorialCalculator()"

    def calculate(self, n: int) -> int:
        if not isinstance(n, int):
            raise TypeError("Input must be an integer.")
        if n < 0:
            raise ValueError("Negative input not allowed")

        result = 1
        for i in range(2, n + 1):
            result *= i
        return result
```

```python
# Step 4B: Backend Setup

import os
import sys
import subprocess

# Base directory
BASE_DIR = "factorial_project"

# -----------------------------
# 1. Create project structure
# -----------------------------
DIRS = [
    f"{BASE_DIR}/src",
    f"{BASE_DIR}/tests",
    f"{BASE_DIR}/backend",
]

for d in DIRS:
    os.makedirs(d, exist_ok=True)

# -----------------------------
# 2. Create files
# -----------------------------
files = {
    f"{BASE_DIR}/src/factorial.py": "# factorial logic",
    f"{BASE_DIR}/tests/test_factorial.py": "# test cases",
    f"{BASE_DIR}/backend/main.py": "# fastapi app",
    f"{BASE_DIR}/requirements.txt": "",
    f"{BASE_DIR}/README.md": "",
}

for path, content in files.items():
    with open(path, "w") as f:
        f.write(content)

# -----------------------------
# 3. Setup backend environment
# -----------------------------
backend_dir = os.path.join(BASE_DIR, "backend")

venv_path = os.path.join(backend_dir, "venv")
subprocess.run([sys.executable, "-m", "venv", venv_path], check=True)

# Activation instructions
if sys.platform == "win32":
    activate_script = os.path.join(venv_path, "Scripts", "activate")
else:
    activate_script = os.path.join(venv_path, "bin", "activate")

print(f"To activate the virtual environment, run: source {activate_script}")

# -----------------------------
# 4. Install dependencies (simulated)
# -----------------------------
dependencies = ["fastapi", "uvicorn", "python-multipart"]

print("\nInstalling dependencies:")
for dep in dependencies:
    print(f"Installing {dep}...")

# -----------------------------
# 5. Print FULL project tree (clean like PDF)
# -----------------------------
print("\nProject Structure:\n")

for root, dirs, files in os.walk(BASE_DIR):
    # Hide venv
    dirs[:] = [d for d in dirs if d != "venv"]

    level = root.replace(BASE_DIR, "").count(os.sep)
    indent = " " * 4 * level

    print(f"{indent}{os.path.basename(root)}/")

    for file in files:
        print(f"{indent}    {file}")
```
